自己回帰係数と偏自己回帰係数は、時系列データにおける過去の値と現在の値との関係を表すという点では似ていますが、**「中間のラグ（時間差）の影響を考慮するかしないか」**という点で数学的な定義が根本的に異なります。

### 1. 自己回帰係数（Autocorrelation Coefficient, ACF）

自己回帰係数（または自己相関係数）は、現在の時点 $t$ の値 $y_t$ と、特定のラグ $k$ だけ離れた過去の値 $y_{t-k}$ との間の**単純な相関**を表します。これは、**中間のラグ（$t-1, t-2, \dots, t-k+1$）をすべて含めた形での全体的な相関関係**を測定します。

数学的には、ラグ $k$ の自己回帰係数 $\rho_k$ は次のように定義されます（定常過程を仮定し、平均値を $\mu$ とする）:

$$
\rho_k = \frac{\text{Cov}(y_t, y_{t-k})}{\text{Var}(y_t)} = \frac{E[(y_t - \mu)(y_{t-k} - \mu)]}{E[(y_t - \mu)^2]}
$$

これは、通常の相関係数の時系列データ版であり、**直接的および間接的な**影響の両方を含んでいます。

### 2. 偏自己回帰係数（Partial Autocorrelation Coefficient, PACF）

偏自己回帰係数は、現在の時点 $t$ の値 $y_t$ と、特定のラグ $k$ だけ離れた過去の値 $y_{t-k}$ との間の**直接的な相関**を表します。これは、**中間のラグ（$t-1, t-2, \dots, t-k+1$）の影響をすべて取り除いた（統計的に調整した）上での相関関係**です。

偏自己回帰係数 $\phi_{k,k}$ は、次数 $k$ の自己回帰（AR($k$)）モデルを当てはめた際の、最後の項（ラグ $k$ の項）の係数として定義されます。

$$
y_t = c + \phi_{k,1}y_{t-1} + \phi_{k,2}y_{t-2} + \dots + \phi_{k,k}y_{t-k} + \epsilon_t
$$

このモデルにおいて、$\phi_{k,k}$ がラグ $k$ の偏自己回帰係数です。これは、ラグ $1$ から $k-1$ までの変数をコントロール変数として、純粋な $y_t$ と $y_{t-k}$ の関係を偏回帰分析的に抽出したものです。これがユールウォーカー方程式の話につながります
