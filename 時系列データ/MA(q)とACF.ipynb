ÊâøÁü•„ÅÑ„Åü„Åó„Åæ„Åó„Åü„ÄÇAR(p)„É¢„Éá„É´„Å®PACF„ÅÆÈñ¢‰øÇ„ÇíÁêÜËß£„Åï„Çå„Çã„Å®„ÄÅ‰ªäÂõû„ÅÆ**MA(q)„É¢„Éá„É´„Å®ACFÔºàËá™Â∑±Áõ∏Èñ¢‰øÇÊï∞Ôºâ**„ÅÆÈñ¢‰øÇ„ÅØ„ÄÅ„Åæ„Åï„Å´„Åù„ÅÆÈè°ÂÜô„ÅóÔºà„Éá„É•„Ç¢„É´Ôºâ„ÅÆ„Çà„ÅÜ„ÅßÈùûÂ∏∏„Å´ÁêÜËß£„Åó„ÇÑ„Åô„Åè„Å™„Çä„Åæ„Åô„ÄÇ

MA(q)„É¢„Éá„É´„ÅÆÊ¨°Êï∞ $q$ „ÇíË¶ã„Å§„Åë„ÇãÈçµ„ÅØ„ÄÅ**ACF„Åå„ÄåqÊ¨°„Åß„Ç´„ÉÉ„Éà„Ç™„ÉïÔºàcut offÔºâ„Åô„Çã„Äç**„Å®„ÅÑ„ÅÜÊÄßË≥™„Å´„ÅÇ„Çä„Åæ„Åô„ÄÇ

### 1. MA(q)„É¢„Éá„É´„Å®„ÅØÔºü

MA(q)„É¢„Éá„É´„ÅØ„Äå**Moving Average model**ÔºàÁßªÂãïÂπ≥Âùá„É¢„Éá„É´Ôºâ„Äç„ÅÆÁï•„Åß„ÄÅÊ¨°Êï∞„Åå $q$ „ÅÆ„É¢„Éá„É´„ÅØ‰ª•‰∏ã„ÅÆ„Çà„ÅÜ„Å´Ë°®„Åï„Çå„Åæ„Åô„ÄÇ

$$
Y_t = c + \epsilon_t + \theta_1 \epsilon_{t-1} + \dots + \theta_q \epsilon_{t-q}
$$

* $Y_t$: ÁèæÂú®„ÅÆÂÄ§
* $c$: ÂÆöÊï∞ÔºàÂπ≥ÂùáÂÄ§„Å´Èñ¢ÈÄ£Ôºâ
* $\epsilon_t$: ÁèæÂú®„ÅÆ„Éé„Ç§„Ç∫Ôºà‰∫àÊ∏¨Ë™§Â∑Æ„ÄÅ„Éõ„ÉØ„Ç§„Éà„Éé„Ç§„Ç∫Ôºâ
* $\epsilon_{t-1}, \dots, \epsilon_{t-q}$: **ÈÅéÂéª $q$ ÊúüÂâç„Åæ„Åß„ÅÆ„Éé„Ç§„Ç∫**
* $\theta_1, \dots, \theta_q$: „Éé„Ç§„Ç∫„Å´„Åã„Åã„Çã‰øÇÊï∞Ôºà„Éë„É©„É°„Éº„ÇøÔºâ

* **Áõ¥ÊÑüÁöÑ„Å™ÊÑèÂë≥:**
    * AR„É¢„Éá„É´„Åå„ÄåÈÅéÂéª„ÅÆ**ÂÄ§** ($Y_{t-p}$)„Äç„Å´‰æùÂ≠ò„Åó„Åü„ÅÆ„Å®ÂØæÁÖßÁöÑ„Å´„ÄÅMA„É¢„Éá„É´„ÅØ„ÄåÈÅéÂéª„ÅÆ**„Éé„Ç§„Ç∫Ôºà‰∫àÊ∏¨Ë™§Â∑Æ $\epsilon_{t-q}$Ôºâ**„Äç„Å´‰æùÂ≠ò„Åó„Åæ„Åô„ÄÇ
    * „ÄåÁèæÂú®„ÅÆÂÄ§ ($Y_t$) „ÅØ„ÄÅÁèæÂú®„ÅÆ„Éé„Ç§„Ç∫„Å®„ÄÅ**$q$ ÊúüÂâç„Åæ„Åß„ÅÆÈÅéÂéª„ÅÆ„Éé„Ç§„Ç∫**„ÅÆÈáç„Åø‰ªò„Åç„ÅÆË∂≥„ÅóÁÆó„ÅßÊ±∫„Åæ„Çã„Äç„Å®„ÅÑ„ÅÜ„É¢„Éá„É´„Åß„Åô„ÄÇ

* **ÈáçË¶Å„Å™ÁÇπ:**
    * $Y_t$ „ÅØ„ÄÅ**$q$ ÊúüÂâç**„ÅÆ„Éé„Ç§„Ç∫ $\epsilon_{t-q}$ „Å´„ÅØ**Áõ¥Êé•**ÂΩ±Èüø„ÇíÂèó„Åë„Åæ„Åô„ÄÇ
    * „Åó„Åã„Åó„ÄÅ**$q+1$ ÊúüÂâç**„ÅÆ„Éé„Ç§„Ç∫ $\epsilon_{t-(q+1)}$ „Åã„Çâ„ÅØ**Áõ¥Êé•ÁöÑ„Å™ÂΩ±Èüø„ÇíÂèó„Åë„Åæ„Åõ„Çì**„ÄÇ

### 2. „Å™„ÅúMA(q)„ÅÆÊ¨°Êï∞Ê±∫ÂÆö„Å´ACF„Çí‰Ωø„ÅÜ„ÅÆ„ÅãÔºü

MA(q)„É¢„Éá„É´„ÅÆÂÆöÁæ©„Å®„ÄÅACFÔºàËá™Â∑±Áõ∏Èñ¢‰øÇÊï∞Ôºö$Y_t$ „Å® $Y_{t-k}$ „ÅÆÁõ∏Èñ¢Ôºâ„ÅÆÂÆöÁæ©„ÇíÁµÑ„ÅøÂêà„Çè„Åõ„Çã„Å®„ÄÅAR„Å®PACF„ÅÆÈñ¢‰øÇ„Å®ÂÖ®„ÅèÈÄÜ„ÅÆ„ÄÅÁæé„Åó„ÅÑÈñ¢‰øÇ„ÅåÂ∞é„Åã„Çå„Åæ„Åô„ÄÇ

> **MA(q)„É¢„Éá„É´„ÅÆACF„ÅØ„ÄÅ„É©„Ç∞ $q$ „Åæ„Åß„ÅØÔºàÈÄöÂ∏∏Ôºâ„Çº„É≠„Åß„ÅØ„Å™„ÅÑÂÄ§„ÇíÊåÅ„Å§„Åå„ÄÅ„É©„Ç∞ $q+1$ ‰ª•Èôç„ÅØÔºàÁêÜË´ñ‰∏äÔºâÂÆåÂÖ®„Å´ 0 „Å®„Å™„Çã„ÄÇ**

**„ÄêÁêÜÁî± (MA(1)„É¢„Éá„É´„ÅßÂçòÁ¥îÂåñ)„Äë**
MA(1)„É¢„Éá„É´ $Y_t = \epsilon_t + \theta_1 \epsilon_{t-1}$ „ÅßËÄÉ„Åà„Å¶„Åø„Åæ„Åó„Çá„ÅÜ„ÄÇ

* **„É©„Ç∞1„ÅÆACF ($Y_t$ „Å® $Y_{t-1}$ „ÅÆÁõ∏Èñ¢):**
    * $Y_t = \underline{\epsilon_t} + \theta_1 \underline{\epsilon_{t-1}}$
    * $Y_{t-1} = \underline{\epsilon_{t-1}} + \theta_1 \epsilon_{t-2}$
    * ‰∏°ËÄÖ„Å´„ÅØÂÖ±ÈÄö„ÅÆ„Éé„Ç§„Ç∫È†Ö **$\epsilon_{t-1}$ „ÅåÂê´„Åæ„Çå„Åæ„Åô**„ÄÇ„Åì„ÅÆÈ†Ö„ÇíÈÄö„Åò„Å¶‰∏°ËÄÖ„ÅØÁõ∏Èñ¢„Åô„Çã„Åü„ÇÅ„ÄÅACF(1)„ÅØ **0„Å´„Å™„Çä„Åæ„Åõ„Çì**„ÄÇ

* **„É©„Ç∞2„ÅÆACF ($Y_t$ „Å® $Y_{t-2}$ „ÅÆÁõ∏Èñ¢):**
    * $Y_t = \epsilon_t + \theta_1 \epsilon_{t-1}$ Ôºà‰Ωø„Çè„Çå„Çã„Éé„Ç§„Ç∫„ÅØ $t$ „Å® $t-1$Ôºâ
    * $Y_{t-2} = \epsilon_{t-2} + \theta_1 \epsilon_{t-3}$ Ôºà‰Ωø„Çè„Çå„Çã„Éé„Ç§„Ç∫„ÅØ $t-2$ „Å® $t-3$Ôºâ
    * ‰∏°ËÄÖ„Å´„ÅØ**ÂÖ±ÈÄö„ÅÆ„Éé„Ç§„Ç∫È†Ö„Åå‰∏ÄÂàá„ÅÇ„Çä„Åæ„Åõ„Çì**„ÄÇ
    * „Éõ„ÉØ„Ç§„Éà„Éé„Ç§„Ç∫„ÅÆ‰ªÆÂÆöÔºà$\epsilon_i$ „Å® $\epsilon_j$ „ÅØ $i \ne j$ „Å™„ÇâÁÑ°Áõ∏Èñ¢Ôºâ„Å´„Çà„Çä„ÄÅ‰∏°ËÄÖ„ÅÆÁõ∏Èñ¢ÔºàACF(2)Ôºâ„ÅØ**ÂÆåÂÖ®„Å´ 0** „Å´„Å™„Çä„Åæ„Åô„ÄÇ

* **„É©„Ç∞ $k \ge 2$ „ÅÆACF:**
    * ÂêåÊßò„Å´„ÄÅ„É©„Ç∞„Åå2‰ª•‰∏äÈõ¢„Çå„Çã„Å®ÂÖ±ÈÄö„ÅÆ„Éé„Ç§„Ç∫È†Ö„Åå„Å™„Åè„Å™„Çã„Åü„ÇÅ„ÄÅACF„ÅØÂÖ®„Å¶ **0** „Å´„Å™„Çä„Åæ„Åô„ÄÇ

„Åì„Çå„ÅåMA(q)„É¢„Éá„É´„ÅÆ„Äå**$q$ Ê¨°„ÅÆ„É©„Ç∞„ÅßÊÄ•„Å´ 0 „Å´ËêΩ„Å°„Çã**„ÄçÔºù„Äå**qÊ¨°„Åß„Ç´„ÉÉ„Éà„Ç™„Éï„Åô„ÇãÔºàÊâì„Å°Âàá„Çå„ÇãÔºâ**„Äç„Å®„ÅÑ„ÅÜÊÄßË≥™„ÅÆÁêÜÁî±„Åß„Åô„ÄÇ

### 3. MA(q)„É¢„Éá„É´„ÅÆPACF„ÅØ„Å©„ÅÜ„Å™„Çã„ÅãÔºü

AR(p)„É¢„Éá„É´„ÅÆACF„ÅåÔºàÈñìÊé•ÁöÑ„Å™ÂΩ±Èüø„ÅßÔºâ„ÇÜ„Å£„Åè„ÇäÊ∏õË°∞„Åó„Åü„ÅÆ„ÇíË¶ö„Åà„Å¶„ÅÑ„Åæ„Åô„ÅãÔºü
„Åù„Çå„Å®ÂÖ®„ÅèÂêå„ÅòÁêÜÂ±à„Åß„ÄÅMA(q)„É¢„Éá„É´„ÅÆPACF„ÅØ„ÄÅ**„ÇÜ„Å£„Åè„Çä„Å®ÔºàÊåáÊï∞Èñ¢Êï∞ÁöÑ„Å´„ÄÅ„Åæ„Åü„ÅØÊåØÂãï„Åó„Å™„Åå„ÇâÔºâ0„Å´Âêë„Åã„Å£„Å¶Ê∏õË°∞Ôºàtail offÔºâ**„Åó„Å¶„ÅÑ„Åç„Åæ„Åô„ÄÇ

Ôºà„Åì„Çå„ÅØ„ÄÅMA(q)„É¢„Éá„É´„ÅåAR(‚àû)„É¢„Éá„É´„ÄÅ„Å§„Åæ„ÇäÁÑ°ÈôêÊ¨°„ÅÆAR„É¢„Éá„É´„Å®„Åó„Å¶Êõ∏„ÅçÁõ¥„Åõ„Çã„Åì„Å®„Å®Èñ¢‰øÇ„Åó„Å¶„ÅÑ„Åæ„Åô„ÄÇÔºâ

### üìä ÂÆü‰æãÔºöMA(1)„É¢„Éá„É´„ÅÆACF„Å®PACF

MA(1)„É¢„Éá„É´Ôºà‰æãÔºö$Y_t = \epsilon_t + 0.8 \epsilon_{t-1}$Ôºâ„ÅÆ„Éá„Éº„Çø„ÇíÁîüÊàê„Åó„ÄÅACF„Å®PACF„Çí„Éó„É≠„ÉÉ„Éà„Åó„Å¶„Åø„Åæ„Åó„Çá„ÅÜ„ÄÇ

* **‰∫àÊÉ≥:**
    * **ACF:** MA(**1**)„É¢„Éá„É´„Å™„ÅÆ„Åß„ÄÅ„É©„Ç∞**1**„Åæ„Åß„ÅØÂÄ§„ÇíÊåÅ„Å§„Åå„ÄÅ„É©„Ç∞**2**‰ª•Èôç„Åß„Äå„Ç´„ÉÉ„Éà„Ç™„Éï„Äç„Åó„ÄÅ0„Å´„Å™„Çã„ÅØ„Åö„ÄÇ
    * **PACF:** „ÇÜ„Å£„Åè„Çä„Å®Ê∏õË°∞„Åô„Çã„ÅØ„Åö„ÄÇ





```python
import numpy as np

import matplotlib.pyplot as plt

import statsmodels.api as sm

from statsmodels.tsa.arima_process import ArmaProcess

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf



# Êó•Êú¨Ë™û„Éï„Ç©„É≥„ÉàË®≠ÂÆö

try:

    plt.rcParams['font.family'] = ['Hiragino Sans', 'Yu Gothic', 'Meiryo', 'TakaoPGothic', 'Noto Sans CJK JP', 'sans-serif']

except Exception:

    plt.rcParams['font.family'] = 'sans-serif'



# MA(1)„É¢„Éá„É´„ÅÆÂÆöÁæ©

# Y_t = epsilon_t + 0.8 * epsilon_{t-1}

ar_params = np.array([1]) # ARÈ†Ö„ÅØ„Å™„Åó (AR(0))

ma_params = np.array([1, 0.8]) # statsmodels„Åß„ÅØ [1, theta_1, theta_2] „Å®ÊåáÂÆö

ma_process = ArmaProcess(ar_params, ma_params)



# „Éá„Éº„Çø„ÇíÁîüÊàê

np.random.seed(42)

y = ma_process.generate_sample(nsample=500)



# ACF„Å®PACF„ÅÆ„Ç∞„É©„Éï„ÇíÊèèÁîª

fig, axes = plt.subplots(2, 1, figsize=(10, 8))



# ACF„ÅÆ„Éó„É≠„ÉÉ„Éà

plot_acf(y, ax=axes[0], lags=20, title='ACF (Ëá™Â∑±Áõ∏Èñ¢‰øÇÊï∞) - MA(1)„É¢„Éá„É´')

axes[0].grid(True, linestyle='--', alpha=0.6)



# PACF„ÅÆ„Éó„É≠„ÉÉ„Éà

plot_pacf(y, ax=axes[1], lags=20, title='PACF (ÂÅèËá™Â∑±Áõ∏Èñ¢‰øÇÊï∞) - MA(1)„É¢„Éá„É´', method='ywm')

axes[1].grid(True, linestyle='--', alpha=0.6)



plt.tight_layout()

```

**„Äê„Ç∞„É©„Éï„ÅÆËß£Èáà„Äë**
Ôºà„Ç∞„É©„Éï„ÅÆÈùí„ÅÑÂ∏Ø„ÅØ„Äå‰ø°È†ºÂå∫Èñì„Äç„Åß„ÅôÔºâ

* **‰∏ä„ÅÆACF:** ‰∫àÊÉ≥ÈÄö„Çä„ÄÅ„É©„Ç∞1„Åß„ÅØ‰ø°È†ºÂå∫Èñì„ÇíÂ§ß„Åç„ÅèË∂Ö„Åà„ÇãÂÄ§ÔºàÁõ∏Èñ¢„ÅÇ„ÇäÔºâ„ÇíÊåÅ„Å£„Å¶„ÅÑ„Åæ„Åô„ÄÇ„Åó„Åã„Åó„ÄÅ**„É©„Ç∞2‰ª•Èôç„ÅØÂÖ®„Å¶‰ø°È†ºÂå∫Èñì„ÅÆÂÜÖÂÅ¥**„Å´Âèé„Åæ„Å£„Å¶„Åä„Çä„ÄÅ„ÄåÁµ±Ë®àÁöÑ„Å´0„Äç„Å®„Åø„Å™„Åõ„Åæ„Åô„ÄÇ
    * „Åì„ÅÆACF„ÅÆ„Ç∞„É©„Éï„Åã„Çâ„ÄÅ„Äå**„É©„Ç∞1„Åß„Ç´„ÉÉ„Éà„Ç™„Éï„Åó„Å¶„ÅÑ„Çã**„Äç„Åì„Å®„ÅåÊòéÁ¢∫„Å´Ë™≠„ÅøÂèñ„Çå„Çã„Åü„ÇÅ„ÄÅ„Åì„ÅÆ„Éá„Éº„Çø„ÅØ **MA(1)** „É¢„Éá„É´„ÅßË™¨Êòé„Åô„Çã„ÅÆ„ÅåÂ¶•ÂΩì„Å†„Çç„ÅÜ„ÄÅ„Å®Ê¨°Êï∞ $q=1$ „ÇíÊ±∫ÂÆö„Åô„Çã„Åì„Å®„Åå„Åß„Åç„Åæ„Åô„ÄÇ

* **‰∏ã„ÅÆPACF:** ‰∫àÊÉ≥ÈÄö„Çä„ÄÅ„É©„Ç∞1, 2, 3...„Å®ÈÄ≤„ÇÄ„Å´„Å§„Çå„Å¶„ÄÅ**„ÇÜ„Å£„Åè„Çä„Å®0„Å´Ëøë„Å•„ÅÑ„Å¶**ÔºàÊåØÂãï„Åó„Å™„Åå„ÇâÊ∏õË°∞„Åó„Å¶Ôºâ„ÅÑ„Åæ„Åô„ÄÇ

### üéØ „Åæ„Å®„ÇÅÔºöAR(p) „Å® MA(q) „ÅÆË¶ãÂàÜ„ÅëÊñπ

ÊôÇÁ≥ªÂàó„Éá„Éº„ÇøÔºà„Ç≥„É¨„É≠„Ç∞„É©„É†Ôºâ„ÇíË¶ã„Å¶„É¢„Éá„É´„ÇíË≠òÂà•„Åô„ÇãÂü∫Êú¨ÁöÑ„Å™ÊñπÊ≥ï„ÅØ„ÄÅ„Åì„ÅÆACF„Å®PACF„ÅÆ„Äå„Ç´„ÉÉ„Éà„Ç™„Éï„Äç„Å®„ÄåÊ∏õË°∞„Äç„ÅÆ„Éë„Çø„Éº„É≥„ÇíË¶ã„Çã„Åì„Å®„Åß„Åô„ÄÇ

| „É¢„Éá„É´ | **ACF**ÔºàËá™Â∑±Áõ∏Èñ¢‰øÇÊï∞Ôºâ | **PACF**ÔºàÂÅèËá™Â∑±Áõ∏Èñ¢‰øÇÊï∞Ôºâ |
| :--- | :--- | :--- |
| **AR(p)** | **Ê∏õË°∞**„Åô„Çã (Tails off) | **pÊ¨°„Åß„Ç´„ÉÉ„Éà„Ç™„Éï** (Cuts off at lag p) |
| **MA(q)** | **qÊ¨°„Åß„Ç´„ÉÉ„Éà„Ç™„Éï** (Cuts off at lag q) | **Ê∏õË°∞**„Åô„Çã (Tails off) |

„Åì„Çå„Åß„ÄÅAR„Å®MA„Å®„ÅÑ„ÅÜÂü∫Êú¨ÁöÑ„Å™ÊôÇÁ≥ªÂàó„É¢„Éá„É´„ÅÆ„ÄåÂûã„Äç„ÇíË¶ã„Å§„Åë„ÇãÊñπÊ≥ï„ÅåÊèÉ„ÅÑ„Åæ„Åó„Åü„ÄÇ

Ê¨°„ÅØ„ÄÅ„Åì„Çå„Çâ2„Å§„ÇíÁµÑ„ÅøÂêà„Çè„Åõ„Åü„Äå**ARMA(p,q)„É¢„Éá„É´**„Äç„Å´„Å§„ÅÑ„Å¶Ëß£Ë™¨„Åô„Çã„ÅÆ„ÅØ„ÅÑ„Åã„Åå„Åß„Åó„Çá„ÅÜ„ÅãÔºü

Á¥†Êô¥„Çâ„Åó„ÅÑÁÇπ„Å´Ê∞ó„Å•„Åã„Çå„Åæ„Åó„Åü„ÄÇÊ∑∑‰π±„Åó„ÇÑ„Åô„ÅÑ„Éù„Ç§„É≥„Éà„Åß„Åô„ÄÇ

ÁßÅ„Åü„Å°„Åå„Ç∞„É©„ÉïÔºà„Ç≥„É¨„É≠„Ç∞„É©„É†Ôºâ„Åß„Éó„É≠„ÉÉ„Éà„Åó„Å¶Ë¶ã„Å¶„ÅÑ„Çã„ÅÆ„ÅØ„ÄÅ„ÅÇ„Åè„Åæ„Åß
„Äå**$Y_t$ÔºàÁèæÂú®„ÅÆÂÄ§Ôºâ„Å® $Y_{t-k}$ÔºàkÊúüÂâç„ÅÆÂÄ§Ôºâ„ÅÆÁõ∏Èñ¢**„Äç
„Åß„Åô„ÄÇ

---

„Åß„ÅØ„ÄÅ„Å™„Åú $\epsilon$Ôºà„Ç§„Éó„Ç∑„É≠„É≥Ôºö„Éé„Ç§„Ç∫Ôºâ„ÅÆË©±„ÅåÂá∫„Å¶„Åç„Åü„Åã„Å®„ÅÑ„ÅÜ„Å®„ÄÅ

* **ACF (Ëá™Â∑±Áõ∏Èñ¢‰øÇÊï∞)**:
