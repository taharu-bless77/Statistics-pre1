## 🎲 幾何分布（失敗回数）の定義

確率変数 $Y$ は「**初めて成功するまでの失敗回数**」と定義します。

* **試行の前提:** ベルヌーイ試行（成功/失敗の2通りの結果）を独立に繰り返し行います。
    * 成功確率: $p$
    * 失敗確率: $q = 1 - p$
* **実現値の集合:** $Y$ の取りうる値は $k = 0, 1, 2, 3, \ldots$ の非負の整数です。
    * $Y=0$: 1回目の試行で成功した（失敗は0回）。
    * $Y=k$: $k$ 回失敗した後、$(k+1)$ 回目の試行で初めて成功した。
* **確率質量関数 (PMF):** $Y=k$ となる確率 $P(Y=k)$ は、最初の $k$ 回が失敗し、次の 1 回が成功する確率なので、
    $$P(Y=k) = \underbrace{q \cdot q \cdots q}_{k \text{回}} \cdot p = q^k p \quad \text{for } k = 0, 1, 2, \ldots$$

## 📈 累積確率関数 $P(Y \ge k)$ の導出

幾何分布の無記憶性を証明するために必要なのが、**「失敗回数が $k$ 回以上となる確率」**、すなわち $P(Y \ge k)$ です。

$Y \ge k$ という事象は、「**最初の $k$ 回の試行がすべて失敗である**」という事象と等価です。なぜなら、$Y$ は**初めて成功するまでの失敗回数**だからです。

* もし最初の $k$ 回がすべて失敗ならば、初めて成功するのは $(k+1)$ 回目以降なので、$Y$ は $k$ 以上の値（$k, k+1, k+2, \ldots$）をとります。
* 逆に、$Y \ge k$ であれば、成功するまでに少なくとも $k$ 回失敗しているため、最初の $k$ 回はすべて失敗だったことになります。

したがって、独立な試行の確率の積として、$P(Y \ge k)$ は以下のように求められます。

$$\begin{aligned}
P(Y \ge k) &= P(\text{1回目失敗} \cap \text{2回目失敗} \cap \cdots \cap \text{k回目失敗}) \\
&= P(\text{1回目失敗}) \cdot P(\text{2回目失敗}) \cdots P(\text{k回目失敗}) \\
&= \underbrace{q \cdot q \cdots q}_{k \text{個}} \\
&= q^k
\end{aligned}$$

このように、**累積確率関数**はシンプルに $P(Y \ge k) = q^k$ となります。

## ✨ 無記憶性の証明（再掲）

**無記憶性**とは、過去の履歴が未来の確率に影響を与えないという性質です。

### 1. 無記憶性の式
「すでに $m$ 回失敗している」という条件の下で、「さらに $n$ 回以上失敗する」（合計 $m+n$ 回以上失敗する）確率が、「最初から $n$ 回以上失敗する」確率と等しいことを示します。
$$P(Y \ge m+n \mid Y \ge m) = P(Y \ge n)$$

### 2. 左辺の計算（条件付き確率の定義）
$$P(Y \ge m+n \mid Y \ge m) = \frac{P((Y \ge m+n) \cap (Y \ge m))}{P(Y \ge m)}$$

### 3. 分子の整理
$Y \ge m+n$ という事象は、自動的に $Y \ge m$ という事象を含みます。したがって、2つの事象の共通部分は $Y \ge m+n$ です。
$$ \frac{P(Y \ge m+n)}{P(Y \ge m)}$$

### 4. 累積確率の公式 $P(Y \ge k) = q^k$ を代入
$$ \frac{q^{m+n}}{q^m}$$

### 5. 計算の実行
指数の法則により、
$$\frac{q^{m+n}}{q^m} = q^{(m+n) - m} = q^n$$

### 6. 右辺との比較
右辺 $P(Y \ge n)$ も、累積確率の公式より $q^n$ です。

$$\text{左辺} = q^n$$
$$\text{右辺} = q^n$$

$$\text{左辺} = \text{右辺}$$

したがって、**幾何分布（失敗回数 $Y$）は無記憶性を持つ**ことが証明されました。これは、過去の試行で何回失敗していようと、次に成功するまでの追加の失敗回数の確率は、**一から試行を始める場合と全く同じ**であることを示しています。 
    $$\frac{q^{m+n}}{q^m} = q^{(m+n)-m} = q^n$$

5.  **右辺との比較**
    右辺 $P(Y \ge n)$ は、累積確率の公式より $q^n$ です。
    $$q^n = P(Y \ge n)$$

$$\text{左辺} = q^n = \text{右辺}$$

したがって、**幾何分布（初めて成功するまでの失敗回数 $Y$）は無記憶性を持ちます**。
