---

## 1. 確率変数 $X$ の基本と $U$ の定義

### 確率変数 $X$ の基本

**確率変数 $X$** は、試行の結果によって値が決まる変数です（ここでは**連続型**を前提とします）。$X$ の分布は以下の2つの関数で特徴づけられます。

* **確率密度関数 (PDF)**: $f_X(x)$
    * $f_X(x) \ge 0$ であり、全範囲で積分すると $1$ になります。
    * $X$ が $x$ の付近の値を取る「相対的な起こりやすさ」を示します。

* **累積分布関数 (CDF)**: $F_X(x)$
    * $F_X(x) = P(X \le x)$ で定義され、$X$ の値が $x$ 以下になる確率を示します。
    * PDFを $-\infty$ から $x$ まで積分したものです: $F_X(x) = \int_{-\infty}^{x} f_X(t) dt$。
    * 値域は $[0, 1]$ であり、単調非減少関数です。

---

### $U$ とは何か

**$U$** は、**一様乱数**、または**一様分布に従う確率変数**を指します。

* **分布**: 区間 $(0, 1)$ 上の**一様分布 ($\mathrm{U}(0, 1)$)** に従います。
* **特徴**: 区間 $(0, 1)$ の中のどの値も等しい確率で出現します。
* **PDF**: $f_U(u) = 1 \quad (0 < u < 1)$
* **CDF**: $F_U(u) = u \quad (0 < u < 1)$

**$U$ は、コンピュータで最も簡単に生成できる基本的な乱数です。**

---

## 2. 確率積分変換と変数の関係

### 確率積分変換 (Probability Integral Transform)

**主張**:
任意の連続型確率変数 $X$ をその累積分布関数 $F_X$ に通して変換した新しい確率変数 $Y$ は、必ず区間 $(0, 1)$ 上の**一様分布**に従います。

$$\text{変換}: \quad \mathbf{Y = F_X(X)}$$

| 変数 | 役割 | 従う分布 | 変数の関係 |
| :--- | :--- | :--- | :--- |
| **$X$** | 元の変数 | 任意の連続分布 ($f_X(x)$) | 変換の**入力** |
| **$Y$** | 変換後の変数 | **一様分布 $\mathrm{U}(0, 1)$** | 変換の**出力** ($Y$ は $U$ と同じ分布になる) |

### 逆関数法 (Inverse Transform Sampling)

**主張**:
区間 $(0, 1)$ 上の一様乱数 $U$ を、目的の分布の累積分布関数 $F_X$ の逆関数 $F_X^{-1}$ に通して変換した新しい確率変数 $X$ は、必ず**目的の分布**に従います。

$$\text{変換}: \quad \mathbf{X = F_X^{-1}(U)}$$

| 変数 | 役割 | 従う分布 | 変数の関係 |
| :--- | :--- | :--- | :--- |
| **$U$** | 一様乱数 | **一様分布 $\mathrm{U}(0, 1)$** | 変換の**入力** |
| **$X$** | 変換後の変数 | 目的の連続分布 ($f_X(x)$) | 変換の**出力** |

### 関係のまとめ

確率積分変換と逆関数法は、**確率変数 $X$ のCDF $F_X(x)$ の単調性（1対1対応）** を利用して、任意の分布と一様分布の間を**双方向**で変換する関係にあります。

$$X \xrightarrow{F_X(\cdot)} U$$
$$U \xrightarrow{F_X^{-1}(\cdot)} X$$

この関係は、シミュレーション（モンテカルロ法）において、コンピュータが作りやすい一様乱数 $U$ を、様々な現実の現象を表現する確率変数 $X$ に変換するために不可欠です。

かしこまりました。**確率積分変換**と**逆関数法**の主張の**証明（導出）**を、変数の関係を明確にしながら、順を追ってご説明します。

---

## 1. 確率積分変換の証明 (任意の分布 $\rightarrow$ 一様分布)

**主張:** 連続型確率変数 $X$ の累積分布関数 (CDF) を $F_X(x)$ とするとき、$Y = F_X(X)$ と定義された新しい確率変数 $Y$ は、区間 $(0, 1)$ 上の**一様分布 $\mathrm{U}(0, 1)$** に従う。

### 1-1. 目標設定

$Y$ が一様分布 $\mathrm{U}(0, 1)$ に従うことを示すには、 $Y$ の累積分布関数 $F_Y(y)$ が、一様分布の $\text{CDF}$ である $F_Y(y) = y \quad (0 < y < 1)$ となることを示せばよいです。

### 1-2. 証明 (導出)

1.  **$Y$ の $\text{CDF}$ を定義に従って書き始める**
    $$F_Y(y) = P(Y \le y)$$

2.  **$Y$ の定義を代入する**
    $$F_Y(y) = P(F_X(X) \le y)$$

3.  **逆関数を適用する**
    $F_X(x)$ は連続型確率変数の $\text{CDF}$ であり、単調非減少（ここでは単調増加と仮定）なので逆関数 $F_X^{-1}(u)$ が存在します。両辺に $F_X^{-1}$ を作用させます。
    $$F_Y(y) = P(X \le F_X^{-1}(y))$$
    *ここで $x = F_X^{-1}(y)$ です。*

4.  **$X$ の $\text{CDF}$ の定義に一致させる**
    $\text{CDF}$ の定義 $F_X(x) = P(X \le x)$ を適用します。
    $$P(X \le F_X^{-1}(y)) = F_X(F_X^{-1}(y))$$

5.  **恒等関数となる**
    関数とその逆関数を合成すると恒等関数になるため、
    $$F_X(F_X^{-1}(y)) = y$$

したがって、 $0 < y < 1$ において、
$$F_Y(y) = y$$
となり、**$Y$ は $\mathrm{U}(0, 1)$ に従う**ことが証明されました。

## 2. 逆関数法の証明 (一様分布 $\rightarrow$ 任意の分布)

**主張:** 区間 $(0, 1)$ 上の一様乱数を $U$ とするとき、$X = F_X^{-1}(U)$ と定義された新しい確率変数 $X$ は、累積分布関数 $F_X(x)$ で表される**目的の分布**に従う。

### 2-1. 目標設定

$X$ が目的の分布に従うことを示すには、 $X$ の累積分布関数 $F_X'(x)$ が、もともと設定した $F_X(x)$ に一致することを示せばよいです。

$$F_X'(x) = P(X \le x)$$

### 2-2. 証明 (導出)

1.  **$X$ の $\text{CDF}$ を定義に従って書き始める**
    $$F_X'(x) = P(X \le x)$$

2.  **$X$ の定義を代入する**
    $$F_X'(x) = P(F_X^{-1}(U) \le x)$$

3.  **逆関数を適用する**
    両辺に $F_X$ を作用させます（$F_X$ は逆関数 $F_X^{-1}$ の逆関数）。
    $$P(F_X^{-1}(U) \le x) = P(U \le F_X(x))$$

4.  **$U$ の $\text{CDF}$ の定義に一致させる**
    $U$ は $\mathrm{U}(0, 1)$ に従うため、その $\text{CDF}$ は $F_U(u) = u$ です。したがって、$u$ の代わりに $F_X(x)$ を代入すると、
    $$P(U \le F_X(x)) = F_U(F_X(x)) = F_X(x)$$

したがって、
$$F_X'(x) = F_X(x)$$
となり、**$X$ は目的の累積分布関数 $F_X(x)$ に従う**ことが証明されました。
