### 例題
$$
\begin{cases}
x + y + z = 6 \\
2x - y + z = 3 \\
x + 2y - z = 2
\end{cases}
$$

### 解説

この方程式の拡大係数行列は次の通りです。
$$[A|\mathbf{b}] = \left[ \begin{array}{ccc|c} 1 & 1 & 1 & 6 \\ 2 & -1 & 1 & 3 \\ 1 & 2 & -1 & 2 \end{array} \right]$$

#### **ステップ1: 1列目の掃き出し**

目標は1列目を $\begin{pmatrix} 1 \\ 0 \\ 0 \end{pmatrix}$ にすることです。

* **操作**: $R_2 \to R_2 - 2R_1$ と $R_3 \to R_3 - R_1$ を行います。
* **本質**: この2つの操作は、それぞれ次の**基本行列** $E_1, E_2$ を左から掛けることに相当します。
    $$E_1 = \begin{pmatrix} 1 & 0 & 0 \\ -2 & 1 & 0 \\ 0 & 0 & 1 \end{pmatrix}, \quad E_2 = \begin{pmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ -1 & 0 & 1 \end{pmatrix}$$
* **計算**: $E_2 E_1 [A|\mathbf{b}]$ を計算します。
    $$\begin{pmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ -1 & 0 & 1 \end{pmatrix} \begin{pmatrix} 1 & 0 & 0 \\ -2 & 1 & 0 \\ 0 & 0 & 1 \end{pmatrix} \left[ \begin{array}{ccc|c} 1 & 1 & 1 & 6 \\ 2 & -1 & 1 & 3 \\ 1 & 2 & -1 & 2 \end{array} \right] = \left[ \begin{array}{ccc|c} 1 & 1 & 1 & 6 \\ 0 & -3 & -1 & -9 \\ 0 & 1 & -2 & -4 \end{array} \right]$$

---

#### **ステップ2: 2列目の掃き出し**

目標は2列目を $\begin{pmatrix} 0 \\ 1 \\ 0 \end{pmatrix}$ にすることです。

* **操作**: $R_2 \to -\frac{1}{3}R_2$ を行い、次にその新しい$R_2$を使って $R_1 \to R_1 - R_2$ と $R_3 \to R_3 - R_2$ を行います。
* **本質**: これらは以下の基本行列 $E_3, E_4, E_5$ を順に掛ける操作です。
    $$E_3 = \begin{pmatrix} 1 & 0 & 0 \\ 0 & -1/3 & 0 \\ 0 & 0 & 1 \end{pmatrix}, \quad E_4 = \begin{pmatrix} 1 & -1 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{pmatrix}, \quad E_5 = \begin{pmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & -1 & 1 \end{pmatrix}$$
* **計算**: $E_5 E_4 E_3$ を現在の行列に掛けた結果は次のようになります。
    $$\left[ \begin{array}{ccc|c} 1 & 0 & 2/3 & 3 \\ 0 & 1 & 1/3 & 3 \\ 0 & 0 & -7/3 & -7 \end{array} \right]$$

---

#### **ステップ3: 3列目の掃き出し**

目標は3列目を $\begin{pmatrix} 0 \\ 0 \\ 1 \end{pmatrix}$ にすることです。

* **操作**: $R_3 \to -\frac{3}{7}R_3$ を行い、次にその新しい$R_3$を使って $R_1 \to R_1 - \frac{2}{3}R_3$ と $R_2 \to R_2 - \frac{1}{3}R_3$ を行います。
* **本質**: 対応する基本行列 $E_6, E_7, E_8$ を掛け合わせます。
    $$E_6 = \begin{pmatrix} 1 & 0 & 0 \\ 0 & 1 & 0 \\ 0 & 0 & -3/7 \end{pmatrix}, \quad E_7 = \begin{pmatrix} 1 & 0 & -2/3 \\ 0 & 1 & 0 \\ 0 & 0 & 1 \end{pmatrix}, \quad E_8 = \begin{pmatrix} 1 & 0 & 0 \\ 0 & 1 & -1/3 \\ 0 & 0 & 1 \end{pmatrix}$$
* **計算**: $E_8 E_7 E_6$ を現在の行列に掛けた最終結果は次のようになります。
    $$\left[ \begin{array}{ccc|c} 1 & 0 & 0 & 1 \\ 0 & 1 & 0 & 2 \\ 0 & 0 & 1 & 3 \end{array} \right]$$

### 結論

全ての行基本変形は、基本行列の積 $P = E_8 E_7 \dots E_1$ を元の行列 $[A|\mathbf{b}]$ に左から掛けることと同じです。この操作によって、左辺の係数行列 $A$ は単位行列 $I$ に変換されました。

$$P A = I$$

これは、$P$ が $A$ の逆行列 $A^{-1}$ であることを意味します。そして、右辺の定数ベクトル $\mathbf{b}$ は $P\mathbf{b}$ すなわち $A^{-1}\mathbf{b}$ に変換され、これが解ベクトル $\mathbf{x}$ となります。

最終的に得られた解ベクトルは、
$$\mathbf{x} = \begin{pmatrix} 1 \\ 2 \\ 3 \end{pmatrix}$$
です。
