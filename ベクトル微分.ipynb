# 【数学】二次形式のベクトル微分 `∂(xᵀAx)/∂x = 2Ax` の証明を丁寧に解説

この記事では、機械学習の数式（特に最小二乗法や正則化など）で頻繁に登場する、二次形式のベクトル微分公式の証明を、行列計算の初心者でも理解できるよう可能な限り詳細に解説します。

## ✍️ 1. 準備：関数の構造を「成分」で理解する

まず、証明の主役である関数 $f(\mathbf{x}) = \mathbf{x}^T\mathbf{A}\mathbf{x}$ が、具体的にどのような計算を行っているのかを解剖します。

#### ベクトルと行列の定義
- $\mathbf{x}$: $k$次元の縦ベクトル
$$\mathbf{x} = \begin{pmatrix} x_1 \\ \vdots \\ x_k \end{pmatrix}$$
- $\mathbf{A}$: $k \times k$ の**対称行列** ($A=A^T$, すなわち $a_{ij} = a_{ji}$)

---

### ステップ1.1: 行列とベクトルの積 `Ax` の解剖

まず、カッコの中の $\mathbf{A}\mathbf{x}$ から見ていきましょう。これは、行列 $\mathbf{A}$ とベクトル $\mathbf{x}$ の積であり、結果は新しい縦ベクトルになります。

$$
\mathbf{A}\mathbf{x} =
\begin{pmatrix}
a_{11} & a_{12} & \cdots & a_{1k} \\
a_{21} & a_{22} & \cdots & a_{2k} \\
\vdots & \vdots & \ddots & \vdots \\
a_{k1} & a_{k2} & \cdots & a_{kk}
\end{pmatrix}
\begin{pmatrix} x_1 \\ x_2 \\ \vdots \\ x_k \end{pmatrix}
=
\begin{pmatrix}
a_{11}x_1 + a_{12}x_2 + \dots + a_{1k}x_k \\
a_{21}x_1 + a_{22}x_2 + \dots + a_{2k}x_k \\
\vdots \\
a_{k1}x_1 + a_{k2}x_2 + \dots + a_{kk}x_k
\end{pmatrix}
$$

この結果のベクトルの $i$ 番目の要素は、**行列Aの$i$行目**と**ベクトルx**の内積であり、総和記号$\Sigma$で書くと $\sum_{j=1}^{k} a_{ij}x_j$ となります。

---

### ステップ1.2: 二次形式 `xᵀAx` の解剖

次に、全体 $f(\mathbf{x}) = \mathbf{x}^T(\mathbf{A}\mathbf{x})$ を計算します。これは、横ベクトル $\mathbf{x}^T$ と、先ほど計算した縦ベクトル $\mathbf{A}\mathbf{x}$ の内積です。

$$
f(\mathbf{x}) =
\begin{pmatrix} x_1 & x_2 & \cdots & x_k \end{pmatrix}
\begin{pmatrix}
\sum_{j=1}^{k} a_{1j}x_j \\
\sum_{j=1}^{k} a_{2j}x_j \\
\vdots \\
\sum_{j=1}^{k} a_{kj}x_j
\end{pmatrix}
$$

この計算は、$x_1 \times (\text{1行目の要素}) + x_2 \times (\text{2行目の要素}) + \dots$ となります。
$$f(\mathbf{x}) = x_1 \sum_{j=1}^{k} a_{1j}x_j + x_2 \sum_{j=1}^{k} a_{2j}x_j + \dots = \sum_{i=1}^{k} x_i \left( \sum_{j=1}^{k} a_{ij}x_j \right)$$

最終的に、すべての項をまとめると、おなじみの二重の総和の形になります。
$$f(\mathbf{x}) = \sum_{i=1}^{k} \sum_{j=1}^{k} a_{ij} x_i x_j$$

---

## 🔍 2. 導出：偏微分のステップを厳密に追跡する

**目標**: 勾配ベクトル $\frac{\partial f}{\partial \mathbf{x}}$ を求める。そのために、まず任意の一つの要素 $x_m$ に関する偏微分 $\frac{\partial f}{\partial x_m}$ を計算します。

### ステップ2.1: $f(x)$ を $x_m$ 中心に4つのグループに分解する

$f(\mathbf{x}) = \sum_{i=1}^{k} \sum_{j=1}^{k} a_{ij} x_i x_j$ の全ての項を、$x_m$ との関係で4つのグループに完璧に分類します。

1.  **グループ1: $i=m$ かつ $j=m$ （対角項）**
    $x_m$ が2回掛け合わされる、ただ一つの項です。
    - **項**: $a_{mm}x_m^2$

2.  **グループ2: $i=m$ だが $j \neq m$ （$m$行目の非対角項）**
    $x_m$ と、それ以外の要素 $x_j$ が掛け合わされる項の集まりです。
    - **項**: $\sum_{j \neq m} a_{mj}x_m x_j$

3.  **グループ3: $j=m$ だが $i \neq m$ （$m$列目の非対角項）**
    $x_m$ と、それ以外の要素 $x_i$ が掛け合わされる項の集まりです。
    - **項**: $\sum_{i \neq m} a_{im}x_i x_m$

4.  **グループ4: $i \neq m$ かつ $j \neq m$ （$x_m$ を全く含まない項）**
    $x_m$ とは無関係な項のすべてです。
    - **項**: $\sum_{i \neq m} \sum_{j \neq m} a_{ij}x_i x_j$

---

### ステップ2.2: 各グループを $x_m$ で偏微分する

- **グループ1の微分**:
    $$\frac{\partial}{\partial x_m} (a_{mm}x_m^2) = 2a_{mm}x_m$$
- **グループ2の微分**: $x_m$ 以外は定数と見なすので、
    $$\frac{\partial}{\partial x_m} \left( \sum_{j \neq m} a_{mj}x_m x_j \right) = \sum_{j \neq m} a_{mj}x_j$$
- **グループ3の微分**: 同様に、
    $$\frac{\partial}{\partial x_m} \left( \sum_{i \neq m} a_{im}x_i x_m \right) = \sum_{i \neq m} a_{im}x_i$$
- **グループ4の微分**: $x_m$ を全く含まないので、微分すると **0** になります。

---

### ステップ2.3: 微分結果を結合し、対称性を適用する

すべての微分結果を足し合わせます。
$$\frac{\partial f}{\partial x_m} = 2a_{mm}x_m + \sum_{j \neq m} a_{mj}x_j + \sum_{i \neq m} a_{im}x_i$$

ここで、**Aが対称行列である**という強力な条件 ($a_{im} = a_{mi}$) を使います。3番目の項は、

$$\sum_{i \neq m} a_{im}x_i \quad \xrightarrow{\text{対称性}} \quad \sum_{i \neq m} a_{mi}x_i$$

これは、和を取るための添え字が $i$ であるか $j$ であるかの違いだけで、2番目の項 $\sum_{j \neq m} a_{mj}x_j$ と全く同じ計算をしています。したがって、2番目と3番目の項は等しく、

$$\frac{\partial f}{\partial x_m} = 2a_{mm}x_m + 2\sum_{j \neq m} a_{mj}x_j$$

---

### ステップ2.4: 行列の積の形へ再構築する

微分結果を`2`でくくりだすと、以下のようになります。

$$\frac{\partial f}{\partial x_m} = 2 \left( a_{mm}x_m + \sum_{j \neq m} a_{mj}x_j \right)$$

ここで、カッコ `( )` の中身に注目してください。
- `$a_{mm}x_m$` は、$j=m$ の場合の項です。
- `$\sum_{j \neq m} a_{mj}x_j$` は、$j$ が $m$ 以外のすべての項の和です。

つまり、カッコの中は **$j=m$ の場合と $j \neq m$ の場合を合わせたもの**ですから、これは **$j=1$ から $k$ までのすべての項の総和**に他なりません。

したがって、この式は一つの $\Sigma$ を使って、次のようにシンプルにまとめることができます。

$$\frac{\partial f}{\partial x_m} = 2 \sum_{j=1}^{k} a_{mj}x_j$$

この $\sum_{j=1}^{k} a_{mj}x_j$ とは、行列 $\mathbf{A}$ の **$m$ 行目**のベクトルと、ベクトル $\mathbf{x}$ の内積であり、行列とベクトルの積 $\mathbf{A}\mathbf{x}$ の**$m$ 番目の要素**の定義そのものです。

---

## 🧠 3. 結論：勾配ベクトルと `2Ax`

私たちは、勾配ベクトル $\nabla f(\mathbf{x})$ の任意の $m$ 番目の要素が、ベクトル $2\mathbf{A}\mathbf{x}$ の $m$ 番目の要素と等しいことを示しました。
これが全ての要素（$m=1, 2, \dots, k$）について成り立つので、ベクトル全体として等しいことが証明されます。

$$
\frac{\partial(\mathbf{x}^T\mathbf{A}\mathbf{x})}{\partial\mathbf{x}} =
\begin{pmatrix} \partial f / \partial x_1 \\ \vdots \\ \partial f / \partial x_k \end{pmatrix}
=
\begin{pmatrix} 2 (\mathbf{A}\mathbf{x})_1 \\ \vdots \\ 2 (\mathbf{A}\mathbf{x})_k \end{pmatrix}
= 2\mathbf{A}\mathbf{x}
$$

**（証明終）**
