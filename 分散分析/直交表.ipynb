## 📈 効率的な実験の技術：分散分析(ANOVA)から直交表まで

私たちの身の回りには「何が結果に影響を与えているのか？」という問いが溢れています。新開発の肥料AとBは、本当に収穫量に差を生むのか？ ポップコーンの味は、塩、バター、火力のどれが一番効いているのか？

これらの問いに統計的に答えるための強力なツールが「**分散分析 (ANOVA)**」であり、それを最小限のコストで実現する技術が「**実験計画法 (DOE)**」です。

### 1. 分析の基本：分散分析 (ANOVA)

分散分析は、データの「全体のバラツキ（総平方和 $S_T$）」を、要因（原因）ごとに分解する手法です。

#### 1元配置：1つの要因を調べる
最もシンプルな形で、「肥料の種類（A, B, C）」だけが「収穫量」に影響するかを調べます。

* **構造モデル**: $Y_{ij} = \mu + \alpha_i + \epsilon_{ij}$
    （データ = 全体平均 + 水準の効果 + 偶然の誤差）
* **平方和の分解**: $S_T = S_A + S_E$
    （全体のバラツキ = 肥料Aによるバラツキ + 誤差のバラツキ）

**F検定**では、肥料によるバラツキ ($V_A$) が、偶然の誤差によるバラツキ ($V_E$) と比べて「十分に大きいか」を判断します。

#### 2元配置：2つの要因と「交互作用」を調べる
「肥料（A, B）」と「品種（X, Y）」が「収穫量」にどう影響するかを見ます。
ここでの最重要概念が「**交互作用**」です。

* **主効果**: 「肥料AはBより良いか？」「品種XはYより良いか？」
* **交互作用**: 「肥料の**効果**は、品種によって**変わる**か？」
    （例：品種Xには肥料Aが効くが、品種Yには肥料Bが効く）



**分析の鉄則**は、**必ず「交互作用」から先に見る**ことです。もし交互作用が有意（＝組み合わせに意味がある）なら、「肥料Aは常に良い」といった主効果だけの議論は無意味になるからです。

#### 乱塊法：「ノイズ」を減らす賢い工夫
実験には「肥料」のような調べたい要因の他に、「畑の日当たり」のような、結果に影響するが調べたいわけではない**ノイズ（ブロック因子）**が存在します。

**乱塊法**は、このノイズの影響（ブロック平方和 $S_R$）を、あらかじめ誤差 $S_E$ から分離（救出）する手法です。（フィッシャーの3原則の「**局所管理**」の実現）

* **（旧）** $S_T = S_A + S_E(\text{旧})$
* **（新）** $S_T = S_A + S_R + S_E(\text{新})$

ノイズを分離した分、誤差 $S_E(\text{新})$ は小さくなります。これにより、$F = V_A / V_E$ の値は大きくなり、肥料Aのわずかな差でも「有意」と検出しやすくなります（＝**検定力が高まる**）。

### 2. 効率化の追求：実験計画法 (DOE)

要因が3つ、5つ、7つと増えていくと、すべての組み合わせ（総当たり）を試すのは現実的ではありません。

* 2水準の要因が3つ $\rightarrow$ $2^3 = 8$回
* 2水準の要因が7つ $\rightarrow$ $2^7 = 128$回
* 2水準の要因が10個 $\rightarrow$ $2^{10} = 1024$回

**実験計画法 (DOE)** は、「最小の実験回数で、最大の情報を引き出す」ための計画手法であり、そのための「魔法のスケジュール表」が「**直交表**」です。

例えば **L8直交表**（$L_8(2^7)$）は、最大7個の2水準要因の影響を、たった**8回の実験**で評価できる設計図です。

---

### 3. 直交表の「カラクリ」と「交絡」

直交表は、ただランダムに実験を間引いたものではなく、緻密な数学的構造を持っています。

#### 3-1. 直交表の設計図
L8直交表の列は、(-1, +1)の水準で考えると、特定の列（基本列 $C_1, C_2, C_4$）の**「要素ごとの積（掛け算）」**で生成されています。

* $C_3 = C_1 \times C_2$
* $C_5 = C_1 \times C_4$
* $C_6 = C_2 \times C_4$
* $C_7 = C_1 \times C_2 \times C_4$

この「積の関係」こそが、直交表の特性（直交性）を生み出すと同時に、「**交絡**」という落とし穴の原因にもなります。

#### 3-2. 交絡（Confounding）とは？
L8直交表は、本来$2^7=128$回で得られる127個分の情報（主効果、交互作用...）を、たった7つの「席（列）」に**圧縮（折りたたみ）**しています。

結果、1つの席（列）に、複数の効果が**「相席」**することになります。これが**交絡**です。
分析してその席（列）に「効果あり！」と出ても、それが相席している効果のうち、どれの手柄なのか区別がつきません。

#### 3-3. 具体例：ポップコーンの悲劇
ポップコーンの味（結果）に対し、3要因（A:塩, B:バター, C:火力）を調べるとします。
最悪なことに、実験者は以下の割り付けをしました。

* **A (塩)** $\rightarrow$ **列3**
* **B (バター)** $\rightarrow$ **列6**
* **C (火力)** $\rightarrow$ **列5**

この直交表の設計図には「$C_3 \times C_6 = C_5$」という数学的関係が隠れています。
これは、「A (塩) と B (バター) の**交互作用 $A \times B$**」が、「C (火力) の**主効果**」と、**列5で完全に交絡（相席）**してしまうことを意味します。



もし、「塩とバターは両方多いか少ないかだと美味しい（＝ $A \times B$ の交互作用が強い）」という真実があった場合、実験者は「**火力が強いと美味しい**（＝ Cの主効果が強い）」と**100%誤解します**。
なぜなら、分析している列5には、火力の効果と塩バターの交互作用の効果がごちゃ混ぜになって現れるからです。

### 4. 「賢い」実験のための割り付け戦略

交絡は避けられませんが、**コントロール**することができます。
厳密な「割り付けルール」はありませんが、以下の戦略（セオリー）に従って、実験者は「何と何を交絡させるか」を意図的に設計します。

1.  **最優先：知りたい効果同士の交絡を避ける**
    [主効果A], [主効果B], [交互作用 $A \times B$] の3つを知りたいなら、これらが別々の列に割り当てられるようにします。
    （例：A→列1, B→列2 とすれば、$A \times B$ は自動的に列3に入るので、列3には何も割り当てない）

2.  **次点：重要な主効果は「きれいな列」に置く**
    列には「生まれ」があります。列1, 2, 4のような基本列は、他の2因子交互作用が紛れ込みにくい「きれいな列」です。最も重要だと思う主効果は、こうした列に割り当てるのがセオリーです。

3.  **実務：実験のしやすさを考慮する**
    * 列1のパターン: `[----++++]` $\rightarrow$ 変更1回
    * 列4のパターン: `[-+-+-+-+]` $\rightarrow$ 変更7回
    「オーブンの温度」のような変更が大変な要因は、変更回数が少ない**列1**に割り当てると、実験が格段に楽になります。

### 結論

分散分析は要因の効果を明らかにする強力なツールですが、多要因を扱うには実験計画法（直交表）による効率化が不可欠です。
しかし、直交表は「交絡」という構造的な特性を内包しています。その設計図（$C_3 \times C_6 = C_5$ など）を理解し、何（主効果）と何（交互作用）を意図的に交絡させ、何を分離するかを戦略的に設計することこそが、効率的で信頼性の高い結論を導く鍵となります。
