分散分析の体系を、数理統計学的な厳密さと実務的な「軸」の感覚で整理します。
ご指定いただいた添字（因子A：$i$、因子B：$j$、ブロックR：$r$、繰り返し：$k$）を使い、モデル式と分散分析表（ANOVA table）を定義します。

---

## 1. 一元配置分散分析 (One-way ANOVA)

最もシンプルな「グループ間の差」を見るモデルです。

### モデル式

$$y_{ik} = \mu + \alpha_i + \epsilon_{ik}$$

* **制約条件:** $\sum_{i=1}^I \alpha_i = 0$
* **誤差項:** $\epsilon_{ik} \sim N(0, \sigma^2)$ （独立）
* **範囲:** $i=1, \dots, I$ （水準数）、$k=1, \dots, K$ （繰り返し数）

### 分散分析表

| 要因 | 平方和 (SS) | 自由度 (df) | 平均平方 (MS) | F値 |
| --- | --- | --- | --- | --- |
| **因子A** | $SS_A = K \sum_i (\bar{y}_{i.} - \bar{y}_{..})^2$ | $I-1$ | $MS_A = \frac{SS_A}{I-1}$ | $\frac{MS_A}{MS_E}$ |
| **残差(E)** | $SS_E = \sum_i \sum_k (y_{ik} - \bar{y}_{i.})^2$ | $I(K-1)$ | $MS_E = \frac{SS_E}{I(K-1)}$ | - |
| **合計(T)** | $SS_T = \sum_i \sum_k (y_{ik} - \bar{y}_{..})^2$ | $IK-1$ | - | - |

---

## 2. 乱塊法 (Randomized Block Design)

「環境の個体差（ブロック）」を第2の要因として追い出し、純粋な因子の効果を取り出すモデルです。

### モデル式

$$y_{ir} = \mu + \alpha_i + \rho_r + \epsilon_{ir}$$

* **制約条件:** $\sum \alpha_i = 0, \ \sum \rho_r = 0$
* **交互作用:** $\alpha \rho = 0$ （存在しないと仮定）
* **範囲:** $i=1, \dots, I$、 $r=1, \dots, R$ （ブロック数）

### 分散分析表

| 要因 | 平方和 (SS) | 自由度 (df) | 平均平方 (MS) | F値 |
| --- | --- | --- | --- | --- |
| **因子A** | $SS_A = R \sum_i (\bar{y}_{i.} - \bar{y}_{..})^2$ | $I-1$ | $MS_A$ | $\frac{MS_A}{MS_E}$ |
| **ブロックR** | $SS_R = I \sum_r (\bar{y}_{.r} - \bar{y}_{..})^2$ | $R-1$ | $MS_R$ | $\frac{MS_R}{MS_E}$ |
| **誤差(E)** | $SS_E = SS_T - SS_A - SS_R$ | $(I-1)(R-1)$ | $MS_E$ | - |
| **合計(T)** | $SS_T = \sum_i \sum_r (y_{ir} - \bar{y}_{..})^2$ | $IR-1$ | - | - |

---

## 3. 二元配置分散分析：繰り返しあり (Two-way ANOVA with Interaction)

因子Aと因子Bの組み合わせによる相乗効果（交互作用）まで考慮するモデルです。

### モデル式

$$y_{ijk} = \mu + \alpha_i + \beta_j + \gamma_{ij} + \epsilon_{ijk}$$

* **制約条件:** $\sum \alpha_i = 0, \ \sum \beta_j = 0, \ \sum_i \gamma_{ij} = 0, \ \sum_j \gamma_{ij} = 0$
* **範囲:** $i=1, \dots, I$、 $j=1, \dots, J$、 $k=1, \dots, K$

### 分散分析表

| 要因 | 平方和 (SS) | 自由度 (df) | 平均平方 (MS) | F値 |
| --- | --- | --- | --- | --- |
| **因子A** | $SS_A = JK \sum_i (\bar{y}_{i..} - \bar{y}_{...})^2$ | $I-1$ | $MS_A$ | $\frac{MS_A}{MS_E}$ |
| **因子B** | $SS_B = IK \sum_j (\bar{y}_{.j.} - \bar{y}_{...})^2$ | $J-1$ | $MS_B$ | $\frac{MS_B}{MS_E}$ |
| **交互作用AB** | $SS_{AB} = K \sum_{i,j} (\bar{y}_{ij.} - \bar{y}_{i..} - \bar{y}_{.j.} + \bar{y}_{...})^2$ | $(I-1)(J-1)$ | $MS_{AB}$ | $\frac{MS_{AB}}{MS_E}$ |
| **誤差(E)** | $SS_E = \sum_{i,j,k} (y_{ijk} - \bar{y}_{ij.})^2$ | $IJ(K-1)$ | $MS_E = \sigma^2$ | - |
| **合計(T)** | $SS_T = \sum_{i,j,k} (y_{ijk} - \bar{y}_{...})^2$ | $IJK-1$ | - | - |

---

## 4. 不釣り合い型 (Unbalanced Data) への対応

各水準のデータ数 $K_{ij}$ が異なる場合、平方和の「直交性」が崩れます。このため、どの変動を先に評価するかによって結果が変わるため、以下の平方和（SS）を選択する必要があります。

### モデル式

基本は二元配置と同じですが、集計時の重みが変わります。

### 平方和の選択基準

* **Type I (Sequential SS):** モデルに投入した順に評価。
* $A \rightarrow B | A \rightarrow AB | A,B$。因子の投入順に意味がある場合（階層モデル等）。


* **Type II (Hierarchical SS):** 交互作用を無視して主効果を評価。
* 交互作用がないと確信できる場合に強力。


* **Type III (Marginal SS):** 他のすべての要因を考慮した後の「純粋な寄与」を評価。
* **実務上の標準。** 欠損値がある場合や、サンプル数がバラバラな実験データで最もよく使われる。



---

### 数理的な総括：なぜ「自由度」で割るのか

あなたが補題の計算で格闘してきた通り、全ての平均平方 ($MS$) は期待値をとると **「$\sigma^2$ + (シグナル成分)」** の形になります。

* 一元配置の $df = N - I$
* 二元配置の $df = IJ(K-1)$

これらは、テンソル空間において「平均（重心）」によって拘束された次元を差し引いた、純粋なノイズ（自由な動き）の次元数を表しています。この自由度で割ることで、全ての要因を「$\sigma^2$」という共通の物差しで比較できるようになるわけです。

実務で Python (statsmodels) 等を使う際は、この不釣り合い型の `typ=3` 指定が、デザイン行列の逆行列を解いている操作に相当することを意識すると、線形代数との繋がりがより深まります。
