院試やデータサイエンスの現場で、不均衡データ（各水準のサンプル数 $n_i$ がバラバラな状態）を扱う際の「数学的な詳細定義の欠如」は、致命的な混乱を招きます。補題 15.1 を単なる公式としてではなく、**「対象となる統計量の定義（期待値・分散）」と「補題の変数」を 1 対 1 で対応させたマッピング**として解体しましょう。

まずは、思考のベースとなる補題の定義を厳密に固定します。

---

### 1. 補題 15.1 の厳密な定義（数学的プロトコル）

確率変数 $X_1, X_2, \dots, X_n$ が互いに独立であるとき、以下の等式が成立します。

$$E\left[ \sum_{t=1}^n (X_t - \bar{X})^2 \right] = \sum_{t=1}^n (\mu_t - \bar{\mu})^2 + (n-1)\sigma_X^2 \quad \dots (\text{Lemma 15.1})$$

この式を適用するにあたって、以下の 4 点を定義しきることが「行間を埋める」作業のすべてです。

1. **対象 $X_t$**: 今、分析のテーブルに乗せている具体的な統計量は何か？
2. **個数 $n$**: その統計量はいくつ並んでいるか？
3. **期待値 $\mu_t, \bar{\mu}$**: その統計量の母平均と、その平均は何か？
4. **分散 $\sigma_X^2$**: **ここが最重要。** 投入した $X_t$ **自身**の分散はいくらか？

---

### 2. 水準内平方和 $SS_W$ への適用（不均衡データ版）

各水準 $i$ ($i=1 \dots a$) において、個別のデータ $y_{ij}$ ($j=1 \dots n_i$) を対象にします。

#### 【マッピング表：水準 $i$ 内部の計算】

| 補題の変数 | 今見ている対象（水準 $i$ 内）の定義 | 数学的根拠 |
| --- | --- | --- |
| **対象 $X_t$** | **$y_{ij}$** ($j=1, \dots, n_i$) | 個別の観測レコード |
| **個数 $n$** | **$n_i$** | 第 $i$ 水準のサンプル数 |
| **期待値 $\mu_t$** | **$\mu + \alpha_i$** | モデルの定義式 $y_{ij} = \mu + \alpha_i + \epsilon_{ij}$ より |
| **平均期待値 $\bar{\mu}$** | **$\mu + \alpha_i$** | 同一水準内では期待値が一定のため |
| **分散 $\sigma_X^2$** | **$\sigma^2$** | 誤差項 $\epsilon_{ij}$ の分散 |

#### 【導出プロセス】

1. **補題への代入**:
水準 $i$ の偏差平方和 $SS_{W,i}$ の期待値は、$\mu_t - \bar{\mu} = 0$ となるため：

$$E[SS_{W,i}] = E\left[ \sum_{j=1}^{n_i} (y_{ij} - \bar{y}_{i.})^2 \right] = 0 + (n_i - 1)\sigma^2$$


2. **全水準の合算**:
すべての水準について和をとります。

$$E[SS_W] = \sum_{i=1}^a E[SS_{W,i}] = \sum_{i=1}^a (n_i - 1)\sigma^2 = (N - a)\sigma^2$$



※ ここで全サンプル数 $N = \sum n_i$。

---

### 3. 水準間平方和 $SS_B$ への適用（不均衡データ版）

不均衡データにおいて、$SS_B$ の定義は **$\sum n_i (\bar{y}_{i.} - \bar{y}_{..})^2$** となります。これは、各水準平均 $\bar{y}_{i.}$ を「データ」と見なした**重み付き**の偏差平方和です。

#### 【マッピング表：水準間の計算】

ここでの「対象」は、各グループの平均値 $\bar{y}_{i.}$ です。

| 補題の変数 | 今見ている対象（水準間）の定義 | 数学的根拠 |
| --- | --- | --- |
| **対象 $X_t$** | **$\bar{y}_{i.}$** ($i=1, \dots, a$) | グループごとの集計値 |
| **個数 $n$** | **$a$** | 水準（グループ）の数 |
| **期待値 $\mu_t$** | **$\mu + \alpha_i$** | $E[\bar{y}_{i.}] = E[\mu + \alpha_i + \bar{\epsilon}_{i.}]$ |
| **平均期待値 $\bar{\mu}$** | **$\mu$** | 制約条件 $\sum n_i \alpha_i = 0$ を課した場合の重み付き平均 |
| **分散 $\sigma_X^2$** | **$\sigma^2 / n_i$** | **ここが等分散ではないため、通常の補題をそのまま使うと誤差が出る** |

#### 【不均衡データ特有の行間埋め：重み付きの処理】

$SS_B$ は単純な補題の形 $\sum (X - \bar{X})^2$ ではなく、前に $n_i$ が付いています。そこで $E[SS_B] = E[\sum n_i \bar{y}_{i.}^2 - N\bar{y}_{..}^2]$ として分解計算します。

1. **第 1 項**: $E[n_i \bar{y}_{i.}^2] = n_i \{ Var(\bar{y}_{i.}) + (E[\bar{y}_{i.}])^2 \} = n_i \{ \frac{\sigma^2}{n_i} + (\mu + \alpha_i)^2 \} = \sigma^2 + n_i(\mu + \alpha_i)^2$
2. **第 1 項の総和**: $\sum_{i=1}^a (\sigma^2 + n_i(\mu + \alpha_i)^2) = a\sigma^2 + N\mu^2 + \sum n_i \alpha_i^2$ （$\sum n_i \alpha_i = 0$ を使用）
3. **第 2 項**: $E[N\bar{y}_{..}^2] = N \{ Var(\bar{y}_{..}) + (E[\bar{y}_{..}])^2 \} = N \{ \frac{\sigma^2}{N} + \mu^2 \} = \sigma^2 + N\mu^2$
4. **差し引き結果**:

$$E[SS_B] = (a\sigma^2 + N\mu^2 + \sum n_i \alpha_i^2) - (\sigma^2 + N\mu^2) = \mathbf{(a-1)\sigma^2 + \sum n_i \alpha_i^2}$$



---

### 4. 結論：平均平方（MS）の期待値構造

これにより、不均衡データにおける分散分析表の期待値（EMS）が「詳細定義」とともに確定します。

| 要因 | 平均平方 $MS = SS/df$ | 期待平均平方 $E[MS]$ |
| --- | --- | --- |
| **水準間** | $MS_B = \frac{SS_B}{a-1}$ | $\sigma^2 + \frac{1}{a-1} \sum n_i \alpha_i^2$ |
| **水準内** | $MS_W = \frac{SS_W}{N-a}$ | $\sigma^2$ |

#### 解決された「詳細定義」のポイント

* **$\sigma^2$ の正体**: 投入した $X_t$ 自体の分散が、平均操作によって $1/n_i$ に目減りし、それを平方和の定義で $n_i$ 倍し直した結果、最終的に $MS$ の世界では「$\sigma^2$」という共通のノイズ単位に揃うことが数学的に保証されました。
* **$F$ 統計量の妥当性**: $F = MS_B / MS_W$ の分母は常に $\sigma^2$、分子は「$\sigma^2 + \text{効果の重み付き和}$」を推定します。だからこそ、サンプル数がバラバラであっても「ノイズを突き抜けた効果」を正しく検定できるのです。

数学的 resolution（解像度）として、このマッピングと導出で「行間」は埋まりましたでしょうか？

もし、この「重み付き制約条件（$\sum n_i \alpha_i = 0$）」を採用しない場合、平均期待値 $\bar{\mu}$ の計算がさらに複雑になりますが、そちらのケースも検討しますか？

「ごめん」なんて言わないでください。ここを自力で完全に埋めようとするその姿勢こそが、数理統計学を「自分の道具」にするための唯一の道ですから。

$E[SS_B] = E\left[ \sum n_i \bar{y}_{i.}^2 - N\bar{y}_{..}^2 \right]$ の展開ですね。
この計算を魔法のようにスッキリ解くための鍵は、統計学の基本等式 **$E[Z^2] = Var(Z) + (E[Z])^2$** を、各統計量に正確にマッピングすることです。

1 つずつ、「行間」を物理的に埋めていきましょう。

## 1. 準備：各統計量の定義を確認する

まず、計算に投入するパーツを整理します。不均衡データ（$n_i$ がバラバラ）における定義です。

* **各群の平均 $\bar{y}_{i.}$**:
* $E[\bar{y}_{i.}] = \mu + \alpha_i$
* $Var(\bar{y}_{i.}) = \frac{\sigma^2}{n_i}$


* **全平均 $\bar{y}_{..}$**:
* $E[\bar{y}_{..}] = \mu$ （※制約条件 $\sum n_i \alpha_i = 0$ を適用した場合）
* $Var(\bar{y}_{..}) = \frac{\sigma^2}{N}$ （※ $N$ は全サンプル数 $\sum n_i$）



---

## 2. 第 1 項の期待値： $E[\sum n_i \bar{y}_{i.}^2]$

シグマの外側に $n_i$ が掛かっているのがポイントです。

$$\begin{aligned}
E\left[ \sum_{i=1}^a n_i \bar{y}_{i.}^2 \right] &= \sum_{i=1}^a n_i E[\bar{y}_{i.}^2] \\
&= \sum_{i=1}^a n_i \left\{ Var(\bar{y}_{i.}) + (E[\bar{y}_{i.}])^2 \right\} \\
&= \sum_{i=1}^a n_i \left\{ \frac{\sigma^2}{n_i} + (\mu + \alpha_i)^2 \right\}
\end{aligned}$$

ここで中身を展開します。

$$\begin{aligned}
&= \sum_{i=1}^a \left( \sigma^2 + n_i(\mu^2 + 2\mu\alpha_i + \alpha_i^2) \right) \\
&= \sum_{i=1}^a \sigma^2 + \sum_{i=1}^a n_i \mu^2 + 2\mu \sum_{i=1}^a n_i \alpha_i + \sum_{i=1}^a n_i \alpha_i^2
\end{aligned}$$

**制約条件 $\sum n_i \alpha_i = 0$** を使うと、$2\mu$ の項が消滅します。
また、$\sum \sigma^2 = a\sigma^2$ 、$\sum n_i \mu^2 = N\mu^2$ ですから：

$$= a\sigma^2 + N\mu^2 + \sum_{i=1}^a n_i \alpha_i^2 \quad \dots (\text{Term A})$$

---

## 3. 第 2 項の期待値： $E[N\bar{y}_{..}^2]$

こちらは全平均についての期待値計算です。

$$\begin{aligned}
E[N\bar{y}_{..}^2] &= N \cdot E[\bar{y}_{..}^2] \\
&= N \left\{ Var(\bar{y}_{..}) + (E[\bar{y}_{..}])^2 \right\} \\
&= N \left\{ \frac{\sigma^2}{N} + \mu^2 \right\} \\
&= \sigma^2 + N\mu^2 \quad \dots (\text{Term B})
\end{aligned}$$

---

## 4. 最終ステップ：引き算

さあ、$SS_B$ の期待値（Term A - Term B）を計算します。

$$\begin{aligned}
E[SS_B] &= (a\sigma^2 + N\mu^2 + \sum n_i \alpha_i^2) - (\sigma^2 + N\mu^2) \\
&= (a - 1)\sigma^2 + \sum_{i=1}^a n_i \alpha_i^2
\end{aligned}$$

---

## 数学的な「納得」のポイント

この計算結果をじっと眺めると、分散分析の「本質」が見えてきます。

1. **$(a-1)\sigma^2$**:
「水準間の差」を測っているつもりでも、そこには必ず **「(水準数 - 1) 個分のノイズ」** が混入してしまうことを示しています。これが $MS_B$ を計算する際に $a-1$ で割らなければならない数学的な理由です。
2. **$\sum n_i \alpha_i^2$**:
不均衡データでは、効果 $\alpha_i$ がサンプル数 $n_i$ で「重み付け」されています。実験データが多いグループのズレほど、平方和を大きく押し上げるという物理的な重みを表しています。

これで「なぜ $SS_B$ の期待値に $a-1$ が現れるのか」、そして「不均衡データでも $MS_B$ は正しく $\sigma^2$ をベースに比較できるのか」、その行間は完全に埋まりました。
