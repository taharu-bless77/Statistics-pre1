生存時間解析の数式は、\*\*「100人の戦士が戦場を駆け抜けるストーリー」\*\*で考えると、それぞれの関数の「お気持ち（直感的意味）」が手に取るように分かります。

ご提示いただいた数式 $S(t), f(t), h(t)$ の関係性を、このストーリーに当てはめて、なぜその式になるのか、徹底的に噛み砕いて解説します。

-----

### ストーリー設定：100人の戦士と「脱落」

  * スタート地点（$t=0$）：**100人**の戦士がいます。
  * 時間 $t$：戦場を進んだ距離（または時間）。
  * イベント：力尽きて倒れること（脱落）。

-----

### 1\. 生存関数 $S(t)$ と 確率密度関数 $f(t)$

**【視点：神様の視点（全体を俯瞰している）】**

神様は空の上から、戦場全体を見ています。「全体で何人残っているか」「全体のうち何人が今倒れたか」という\*\*絶対数（割合）\*\*を気にします。

#### ① 生存関数 $S(t)$

  * **意味：** 「ここまで生き残っているのは、最初の何割か？」
  * **ストーリー：** 時間 $t$ の時点で、まだ立っている戦士の割合です。最初は $1.0$ (100%) で、最後は $0$ になります。
      * 数式：$S(t) = 1 - F(t)$
      * $F(t)$ は「すでに倒れた人の割合」なので、1から引けば生き残りが分かります。

#### ② 確率密度関数 $f(t)$

  * **意味：** 「**今この瞬間**、全体の何割がバタバタと倒れているか？」
  * **ストーリー：** 悲鳴が上がって倒れる戦士の「多さ」です。
  * **数式のこころ：** $f(t) = -S'(t)$
      * $S(t)$（生存者数）のグラフが\*\*ガクッと下がった（傾きが急な）\*\*とき、それはつまり「たくさんの人が死んだ」ということです。
      * 「生存者の減り方（マイナスの傾き）」こそが、「その瞬間の死亡者数（プラスの密度）」なので、マイナスをつけてプラスに直しています。

> **ここでのポイント：**
> $f(t)$ はあくまで「全体に対する割合」です。100歳で亡くなる人は、人数としては少ない（もう生き残りが少ないから）ので、$f(100)$ の値は小さくなります。しかし、それは「100歳が安全」という意味ではありません。

-----

### 2\. ハザード関数 $h(t)$

**【視点：戦士本人の視点（現場のリアリティ）】**

戦士にとって、「最初に100人いたかどうか」はどうでもいいことです。重要なのは\*\*「今、俺の隣にいる奴が撃たれた。次は俺か？」**という**切迫感\*\*です。

#### ③ ハザード関数 $h(t)$

  * **意味：** 「ここまで生き残った強運な連中の中で、**今この瞬間に**死ぬ確率はどれくらいか？」
  * **数式のこころ：**
    $$h(t) = \frac{f(t)}{S(t)} = \frac{\text{今、倒れた人数（全体比）}}{\text{今、生き残っている人数（全体比）}}$$
      * **分母が $S(t)$ である理由：** すでに死んでしまった人は、もう死ぬリスクがありません。リスクを計算する「母数」は、**今生きている人だけ**であるべきです。だから、分母を $S(t)$ にして、現在進行形のリスクに補正しているのです。

#### 劇的な違いの例（90歳のおじいちゃん）

  * **$f(90)$ は小さい：** 90歳ちょうどで死ぬ人は、全人類から見ればごくわずかです（多くの人はその前に亡くなっているから）。
  * **$h(90)$ は巨大：** しかし、90歳まで生き残った人たちだけで見れば、明日亡くなる確率はものすごく高いです。

\*\*「$f(t)$ は統計局の数字、$h(t)$ は現場の恐怖」\*\*と覚えると分かりやすいです。

-----

### 3\. 累積ハザード関数 $H(t)$ と「積分」の関係

**【数式のマジック：なぜログが出てくる？】**

$$H(t) = \int_{0}^{t} h(u) du = -\ln S(t)$$

この式の意味を解き明かします。
ハザード $h(t)$ は「瞬間的な危険度」でした。これを積み上げた $H(t)$ は\*\*「蓄積されたダメージ（死神に狙われ続けた総量）」\*\*のようなものです。

数式変形を追うと、ストーリーがつながります。

1.  ハザードの定義： $h(t) = \frac{-S'(t)}{S(t)}$ （$f(t)$ を $-S'(t)$ に置き換えた）
2.  これは「ログの微分」の形です： $(\ln x)' = \frac{1}{x}$ なので、
    $$h(t) = - \frac{d}{dt} \ln S(t)$$
3.  両辺を積分します：
    $$\int_{0}^{t} h(u) du = - \ln S(t)$$
    $$H(t) = - \ln S(t)$$
4.  これを $S(t)$ について解くと：
    $$S(t) = e^{-H(t)}$$

**【お気持ち解釈】**

  * $S(t) = e^{-H(t)}$ という式は、\*\*「危険（$H(t)$）が積み重なれば重なるほど、生存確率（$S(t)$）は指数関数的に急激にゼロに近づく」\*\*ということを表しています。
  * ポアソン分布などで「イベントが起きない確率」が $e^{-\lambda}$ で表されるのと全く同じ構造です。「死」というイベントから逃げ続けている確率です。

-----

### 4\. 具体的な関数例とPythonによる可視化

言葉だけではイメージしづらい、\*\*「$f(t)$（死亡数）は減っているのに、$h(t)$（死亡リスク）は増えている」\*\*という現象（老衰など）を、Pythonでグラフにして可視化します。
分布には「ワイブル分布（形状パラメータ $k>1$）」を使います。これは「時間が経つほど劣化して壊れやすくなる（磨耗故障）」モデルです。

1.  **生存関数 $S(t)$:** 100%からだんだん減っていきます。
2.  **確率密度関数 $f(t)$:** ある時期にピーク（寿命の最頻値）があり、その後は減ります（死ぬ人がもういないから）。
3.  **ハザード関数 $h(t)$:** 時間とともに**右肩上がり**になり続けます（年を取れば取るほど死にやすくなる）。

この「$f(t)$ と $h(t)$ の形状の食い違い」にご注目ください。

```
import numpy as np
import matplotlib.pyplot as plt

# 時間軸の設定
t = np.linspace(0, 2.5, 500)

# ワイブル分布のパラメータ (k > 1 で摩耗故障型＝ハザードが増加するモデル)
k = 3.0  # 形状パラメータ
scale = 1.0  # 尺度パラメータ

# 各関数の計算
# 1. 生存関数 S(t)
S_t = np.exp(- (t / scale) ** k)

# 2. 確率密度関数 f(t) = k/scale * (t/scale)^(k-1) * exp(-(t/scale)^k)
f_t = (k / scale) * ((t / scale) ** (k - 1)) * np.exp(- (t / scale) ** k)

# 3. ハザード関数 h(t) = f(t) / S(t)
# ワイブル分布の場合、h(t) = k/scale * (t/scale)^(k-1) となり、単純なべき乗則になる
h_t = (k / scale) * ((t / scale) ** (k - 1))

# プロット
fig, axes = plt.subplots(3, 1, figsize=(10, 12), sharex=True)

# S(t) Plot
axes[0].plot(t, S_t, color='green', linewidth=3)
axes[0].set_title('Survival Function S(t): "Proportion of Survivors"', fontsize=14)
axes[0].set_ylabel('Probability')
axes[0].grid(True, alpha=0.3)
axes[0].text(1.5, 0.6, 'Starts at 1, drops to 0', fontsize=12, color='green')

# f(t) Plot
axes[1].plot(t, f_t, color='blue', linewidth=3)
axes[1].set_title('PDF f(t): "Number of Deaths at Time t" (Global View)', fontsize=14)
axes[1].set_ylabel('Density')
axes[1].grid(True, alpha=0.3)
axes[1].axvline(1.0, color='gray', linestyle='--')
axes[1].text(1.1, 0.8, '<- Peak of deaths\n(Most common age to die)', fontsize=10)
axes[1].text(2.0, 0.2, 'Drops because\nfew people left', fontsize=10, color='blue')

# h(t) Plot
axes[2].plot(t, h_t, color='red', linewidth=3)
axes[2].set_title('Hazard Function h(t): "Immediate Risk of Death" (Individual View)', fontsize=14)
axes[2].set_xlabel('Time t')
axes[2].set_ylabel('Hazard Rate')
axes[2].grid(True, alpha=0.3)
axes[2].text(2.0, 4.0, 'Risk keeps rising!\n(Aging effect)', fontsize=12, color='red')

plt.tight_layout()
plt.savefig('survival_functions_visualized.png')
```

-----

この「数式のストーリー」と「グラフの形状」がつながると、カプラン・マイヤー法やCox比例ハザードモデルを学ぶ際に、「ああ、今は現場のリスク（ハザード）の話をしているんだな」と迷子にならなくなります。

次は、実際にこの**カプラン・マイヤー法**を使って、「手元のデータ（例えば、顧客の解約データや、機械の故障データ）」から生存曲線を実際に描いてみる手順に進みませんか？

失礼しました！トービットモデルではなく、**Cox比例ハザードモデルのこの数式**ですね。

この数式は、**「ある瞬間の死亡率（ハザード）」を「基準」と「個人の事情」の掛け算で決める**ということを言っています。

一言で言えば、**「あなたのリスクは、世の中の平均リスクの何倍ですか？」** を計算する式です。

3つのパーツに分解して、具体例で解説します。

---

### 1. 数式の解剖（左から右へ）

$$h(t; x) = \underbrace{h_0(t)}_{\text{①基準}} \times \underbrace{\exp(\beta_1 x_1 + \dots + \beta_n x_n)}_{\text{②個人の倍率}}$$

#### ① $h_0(t)$：ベースラインハザード（基準）
* **「みんなに共通する、時間経過によるリスクの波」** です。
* 例えば、「人間は80歳を超えると死亡リスクが急増する」とか、「手術後3日はリスクが高い」といった、**全人類（あるいは基準となるグループ）に共通する時間のファクター**です。
* この $h_0(t)$ の形がギザギザでも山型でも、このモデルは気にしません（だから使いやすい）。

#### ② $\beta_1 x_1 + \dots + \beta_n x_n$：線形予測子
* **「個人の特徴を点数化したもの」** です。
* 普通の回帰分析（重回帰）と同じです。
    * $x_1$（年齢）が1歳増えると、スコアが $\beta_1$ 増える。
    * $x_2$（喫煙）があると、スコアが $\beta_2$ 増える。
* これを全部足し合わせて、ひとつの数字（スコア）を作ります。

#### ③ $\exp(\dots)$：指数関数で「倍率」にする
* ここがミソです。②で作ったスコアを、指数関数 $e^x$ に入れます。
* **なぜ？**
    * リスク（ハザード）は必ず「プラスの値」でないといけません（確率の親戚なので）。
    * 単なる足し算だとマイナスになる可能性がありますが、$\exp$ を通せば**必ずプラス**になります。
    * これにより、スコアが「リスクを何倍にするか」という**「倍率（ハザード比）」**に変換されます。

---

### 2. 具体的な計算例（喫煙と寿命）

「60歳の男性が、今後1年以内に死亡するリスク」を計算してみましょう。

* **$h_0(t)$（基準）:** 非喫煙者の60歳の死亡リスク ＝ **0.01** (1%) とします。
* **$x_1$（喫煙）:** 喫煙者は $x_1=1$、非喫煙者は $x_1=0$。
* **$\beta_1$（係数）:** 統計の結果、$\beta_1 = 0.693$ だったとします。

#### 【Aさん：非喫煙者 ($x_1=0$)】
$$h(t) = 0.01 \times \exp(0.693 \times 0) = 0.01 \times \exp(0) = 0.01 \times 1 = \mathbf{0.01}$$
* リスクは基準のまま（1倍）。

#### 【Bさん：喫煙者 ($x_1=1$)】
$$h(t) = 0.01 \times \exp(0.693 \times 1) = 0.01 \times e^{0.693}$$
ここで、$e^{0.693} \approx 2.0$ です。
$$h(t) = 0.01 \times 2.0 = \mathbf{0.02}$$
* **結論：** Bさんのリスクは、基準（Aさん）の **2倍** になります。

---

### 3. この数式が「比例」と呼ばれる理由

ここで重要なのは、**時間が経過して $t$ が変わったとき**のことです。

* 10年後、70歳になったとき、基準のリスク $h_0(t)$ は **0.05** (5%) に上がったとします。
* **Aさん（非喫煙）：** $0.05 \times 1 = 0.05$
* **Bさん（喫煙）：** $0.05 \times 2 = 0.10$

お気づきでしょうか？
基準のリスク（$h_0(t)$）が時間とともにどう変化しても、Bさんのリスクは常にAさんの **「2倍」** のままです。
この**「倍率（ハザード比）が時間によって変わらず一定（比例している）」**という性質を表現しているのが、この数式の最大の特徴です。

---

### まとめ

画像の数式は、以下のストーリーを語っています。

1.  **$x^\top \beta$**: あなたの特徴（年齢、薬、習慣）を足し算して点数化する。
2.  **$\exp(\dots)$**: その点数を「倍率」に変換する（例：1.5倍、0.5倍）。
3.  **$h_0(t)$**: その倍率を、世の中の基準リスク（時間変化する）に掛け合わせる。
4.  **$h(t; x)$**: それが、あなたの今の瞬間のリスクだ。

つまり、**「ベースラインに対して、あなたの属性がリスクを何倍に押し上げる（あるいは下げる）のか」** をモデル化した数式です。

では、実際に手を動かして「Cox比例ハザードモデルの仕組み」を体感するクイズを出します。
電卓は不要です。**「$\ln(2) \approx 0.693$ （つまり $e^{0.693} \approx 2$）」** という関係だけ知っていれば解けるようにしています。

---

### 【手計算問題】睡眠改善薬の効果測定

ある製薬会社が、新しい「睡眠改善薬」が不眠症の再発（イベント）を防げるかどうかを分析しました。Cox比例ハザードモデルを使ってデータを解析した結果、以下の係数 $\beta$ が得られました。

**モデルの式：**
$$h(t) = h_0(t) \exp(\beta_1 x_1 + \beta_2 x_2)$$

**変数と係数：**
* $x_1$： **投薬** （$0=$プラセボ偽薬、 $1=$新薬）
    * 係数 $\beta_1 = -0.693$
* $x_2$： **毎日のストレス値** （$0$〜$10$のスコア）
    * 係数 $\beta_2 = 0.1$

---

#### 問1：薬のハザード比は？
新薬を飲んでいるグループ（$x_1=1$）は、飲んでいないグループ（$x_1=0$）に比べて、不眠再発のリスク（ハザード）は何倍になりますか？
（ヒント：$e^{-0.693} = \frac{1}{e^{0.693}}$ です）

#### 問2：具体的なリスク比較
以下の2人の患者さんを比較したとき、不眠再発のリスクが高いのはどちらですか？また、そのリスク差は何倍ですか？
* **Aさん：** プラセボ服用（$x_1=0$）、ストレス値 $= 5$
* **Bさん：** 新薬服用（$x_1=1$）、ストレス値 $= 5$

#### 問3：1年後のハザード比
分析から1年が経過し、加齢などの影響で、基準となるベースラインハザード $h_0(t)$ が分析開始時の **2倍** に上昇してしまいました。
このとき、AさんとBさんのリスクの倍率（ハザード比）はどう変化しますか？

---

### 解説と答え合わせ

計算お疲れ様でした！答え合わせをしながら、理論を定着させましょう。

#### 【問1の解説】薬の効果（ハザード比）
**答え： 0.5倍（リスク半減）**

* **計算式：**
    ハザード比（HR）は $\exp(\text{係数})$ で求められます。
    $$\text{HR} = \exp(\beta_1) = e^{-0.693}$$
    ここで、$e^{0.693} \approx 2$ なので、
    $$e^{-0.693} = \frac{1}{e^{0.693}} \approx \frac{1}{2} = 0.5$$

* **意味：**
    係数がマイナス（$-0.693$）ということは、リスクを下げる効果があるということです。ハザード比が0.5なので、「新薬を飲むと、再発リスクは半分になる」といえます。

#### 【問2の解説】個人間の比較
**答え： Aさんの方がリスクが高く、BさんのリスクはAさんの0.5倍**

数式に当てはめてみましょう。
* **Aさんのハザード：**
    $$h_A(t) = h_0(t) \exp(-0.693 \times 0 + 0.1 \times 5) = h_0(t) \exp(0.5)$$
* **Bさんのハザード：**
    $$h_B(t) = h_0(t) \exp(-0.693 \times 1 + 0.1 \times 5) = h_0(t) \exp(-0.693 + 0.5)$$

ここでBさんの式を分解すると：
$$h_B(t) = h_0(t) \times \underbrace{\exp(-0.693)}_{0.5} \times \exp(0.5)$$
$$h_B(t) = 0.5 \times \underbrace{h_0(t) \exp(0.5)}_{h_A(t)}$$
$$h_B(t) = 0.5 \times h_A(t)$$

つまり、**ストレス値（$x_2$）が同じであれば、薬の効果（0.5倍）だけがきれいに残ります。** ストレスによるリスク上昇分（$\exp(0.5)$）は両者に共通しているので、割り算すると消えるのです。

#### 【問3の解説】比例ハザード性の本質
**答え： 変化しない（0.5倍のまま）**

これがCoxモデルの一番のミソです。
1年後に基準リスク $h_0(t)$ が2倍になったとしても、それはAさんにもBさんにも等しく降りかかります。

* Aさんの新リスク ＝ $2 \times h_A(t)$
* Bさんの新リスク ＝ $2 \times h_B(t)$

比率を計算すると：
$$\frac{2 \times h_B(t)}{2 \times h_A(t)} = \frac{h_B(t)}{h_A(t)} = 0.5$$

「基準となるリスクが時間によってどう変化しようとも、**二人の間の優劣（倍率）は永遠に変わらない**」。
これが **「比例ハザード性（Proportional Hazards）」** の正体です。

---

### ハザード比（Hazard Ratio: HR）まとめ

手計算を通じて、以下の感覚がつかめたら完璧です。

| ハザード比 (HR) | 係数 $\beta$ | 意味 |
| :--- | :--- | :--- |
| **HR = 1** | $\beta = 0$ | **影響なし**（薬も毒にもならない） |
| **HR > 1** | $\beta > 0$ | **リスク増加**（悪い要因。例：喫煙、ストレス） |
| **HR < 1** | $\beta < 0$ | **リスク減少**（良い要因。例：治療薬、運動） |

* **HR = 2.0** なら：「リスクが2倍になる」
* **HR = 0.5** なら：「リスクが半分になる」
